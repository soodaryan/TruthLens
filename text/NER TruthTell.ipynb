```python
from gliner import GLiNER

model = GLiNER.from_pretrained("urchade/gliner_base")
```

```python

news_list = [
    "World Chess Championship 2024 Game 11 Highlights: Gukesh defeats Ding Liren with white pieces",
    "Adobe’s stock drop highlights this fundamental disconnect over AI",
    "Susan Polgar on 18-year-old Gukesh’s win at World Chess Championship: ‘He kept on relentless pressure, Ding self-destructed’",
    "Apple supercharges Voice Memos with AI-powered layer recording support on iPhone 16 Pro series",
    "Increase PMAY-G housing aid, inflation hurting rural home dreams, says Parliamentary Panel"
]

labels = [
    "person", "organization", "location", "date", "time", "event",
    "product", "work_of_art", "law", "percent", "money",
    "quantity", "ordinal", "cardinal", "miscellaneous", "technology", "finance", "policy"
]

for news in news_list:
    print(f"\nOriginal News: {news}")
    entities = model.predict_entities(news, labels)

    # Display the identified entities
    if entities:
        print("Extracted Entities:")
        for entity in entities:
            print(f"{entity['text']} => {entity['label']}")
    else:
        print("No entities were found in this news.")


```
